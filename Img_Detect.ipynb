```python
import cv2
import os
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detectImage(fcascade,clrd_img,scalefactor=1.1):
    img_cpy = clrd_img.copy()
    gimg_cpy = cv2.cvtColor(img_cpy, cv2.COLOR_BGR2GRAY)
    faces = fcascade.detectMultiScale(gimg_cpy, scaleFactor=scalefactor,minNeighbors=2);
    for (x, y, w, h) in faces:     
        cv2.rectangle(img_cpy, (x, y), (x+w, y+h), (0, 255, 0), 2)
    plt.imshow(convertToRGB(img_cpy))
```

```python
test1 = cv2.imread('grp.jpeg')
haar_cascade_classifier = cv2.CascadeClassifier('/home/ns/anaconda3/share/OpenCV/haarcascades/haarcascade_frontalface_alt2.xml')
```
